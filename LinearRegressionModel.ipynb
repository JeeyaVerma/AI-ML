### A) Import important libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### B) Generation of datapoint

```python
X=np.linspace(0,1,100) ## generates data between 0 and 10
X.ndim
```

```python
X.shape
```

```python
X.size
```

```python
### Actual model(m=2,c=1)
Y=2*X+1
```

### C) Plot the datapoints

```python
plt.plot(X,Y,'+')
```

```python
plt.plot(X,Y,'+')
plt.xlabel("input datapoints")
plt.ylabel("output datapoints")
plt.title("random datapoints")
plt.show() #only image will be shown
```

### D) Build a Linear Model

#### 1. Partition your data into training data and test data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y)
```

```python
xtrain.size
```

```python
xtest.size
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2) ##since the data is randomly divided, it divides the data into 20%-80%
```

```python
xtrain.size
```

```python
xtest.size
```

```python
xtest1=xtest.reshape(-1,1)
xtest1.shape ##Linear Regression only takes 2D data
```

#### 2.Linear Model

```python
from sklearn.linear_model import LinearRegression
```

```python
linearmodel=LinearRegression()
```

```python
xtrain=xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
linearmodel.fit(xtrain,ytrain) #fit function gives beta
```

#### 3. Testing of Linear Model

```python
xtest=xtest.reshape(-1,1)
ypred=linearmodel.predict(xtest)
```

```python
ypred
```

```python
ytest
```

#### E) Visualise the actual data and predicted data

```python
plt.plot(xtest,ypred,color='blue') ###for predicted data
plt.show
```

```python
plt.plot(xtrain,ytrain,color='red') ### for actual data
plt.show
```

```python
plt.scatter(xtrain,ytrain,color='red') ### for actual model
plt.plot(xtest,ypred,color='blue') ###for predicted model
plt.show
```
