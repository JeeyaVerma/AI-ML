```python
import numpy as np
students = np.array(["Alice","Bob","Carol","David","Eva"])
marks = np.array([
    [88,92,85],
    [75,78,80],
    [93,88,90],
    [70,72,68],
    [85,90,87]
])

#Basic Statistics
#Normalization (scale to 0-1)
#Topper per subject
#Add 5 grace marks to anyone scoring <75 in any subject.
#compute overall class average after grace marks.
#Identify students whose average is >=90 and label them "Star Performers".
```

```python
# 1. Basic Statistics
print("Basic Statistics:")
print("Mean per subject:", marks.mean(axis=0))
print("Min per subject:", marks.min(axis=0))
print("Max per subject:", marks.max(axis=0))
print("student-wise avg:",np.round(marks.mean(axis=1),2))
```

```python
# 2. Normalization (scale to 0-1)
normalized_marks = (marks - marks.min(axis=0)) / (marks.max(axis=0) - marks.min(axis=0))
print("Normalized Marks (0 to 1):")
print(np.round(normalized_marks, 2))
print()
```

```python
# 3. Topper per subject
import numpy as np
topper_indices = np.argmax(marks,axis=0)
topper_names = students[topper_indices]

subjects=["math","science","english"]
for i in range(3):
    print(f"topper in {subjects[i]}: {topper_names[i]}with {marks[topper_indices[i],i]}")
```

```python
# Step 4: Add 5 Grace Marks to Anyone Scoring <75
# Copy to preserve original
adjusted_marks = marks.copy()
adjusted_marks[adjusted_marks < 75] += 5
print("marks before grace:\n",marks)
print("Adjusted Marks (after grace):\n", adjusted_marks)
```

```python
# 5. Compute overall class average after grace marks
class_average = np.mean(adjusted_marks)
print("Overall class average after grace:", round(class_average, 2))
print()
```

```python
# 6. Identify 'Star Performers' (average >= 90)
student_averages = np.mean(adjusted_marks, axis=1)
star_performers = students[student_averages >= 90]
print("Star Performers (avg >= 90):")
print(star_performers)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#creating dist.
mu,sigma=0,1 #mean=0, std dev=1
x=np.linspace(-4,4,1000)
y=norm.pdf(x,mu,sigma)

#plot
plt.figure(figsize=(10,5))
plt.plot(x,y,label="Normal Distribution")
plt.fill_between(x,y,where=(x>=-1) & (x<=1), alpha=0.3, label = '68% within one row')
plt.fill_between(x,y,where=(x>=-2) & (x<=2), alpha = 0.2,label = '95% within 2 row' )
plt.fill_between(x,y,where=(x>=-3) & (x<=3), alpha = 0.2,label = '99.73% within 3 row' )
plt.title("standard Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("probability density")
plt.legend()
plt.grid()
plt.show()
```

ANYTHING OUTSIDE -3 AND 3 IS CALLED AS AN OUTLIER. SOMETIMES THEY ARE TOO POWERFUL TO IGNORE.

WHEN THE DATA IS NOT NORMALLY DISTRIBUTED THEN WE DON'T USE Z-SCORE

Z-SCORE FAILS IF THERE IS HIGH SKEWEDNESS IN DATA.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew

# generate data
np.random.seed(42)  
Normal_data = np.random.normal(loc=0, scale=1, size=1000)  
pos_skewed = np.random.exponential(scale=1, size=1000)
neg_skewed = -np.random.exponential(scale=1, size=1000)  

#plot
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
for ax, data, title in zip(axs, [Normal_data, pos_skewed, neg_skewed],
                         ['Normal(skew=0)', 'positive skew', 'Negative skew']):
    ax.hist(data, bins=30, color='skyblue', edgecolor='black')
    ax.set_title(f"{title}\nSkewness:{skew(data):.2f}")
plt.tight_layout()
plt.show()
```
